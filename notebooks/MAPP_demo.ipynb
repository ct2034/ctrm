# Simple setup for MAPP in continuous spaces

```python
from ctrm.environment import generate_ins_2d_with_obs_sphere
from ctrm.viz import simple_plot_2d, plot_trm_onestep
from ctrm.roadmap import get_timed_roadmaps_random
from ctrm.planner import PrioritizedPlanning as PP
from ctrm.utils import set_global_seeds

set_global_seeds(0)
```

```python
# generate instance
ins = generate_ins_2d_with_obs_sphere(
    num_agents=2,
    max_speed=0.25,
    rad=0.05,
    obs_num=2,
)

simple_plot_2d(ins, output_size=4)
```

```python
# create timed roadmap
trms = get_timed_roadmaps_random(ins, T=10, num=50)

print("agent 0")
plot_trm_onestep(trms[0], ins)

print("agent 1")
plot_trm_onestep(trms[1], ins)
```

```python
# define planner
planner = PP(ins, trms)

# solve
res = planner.solve()
print(f"elapsed={res.elapsed_planner}")
print(f"makespan={res.maximum_costs}")
print(f"sum-of-costs={res.sum_of_costs}")
    
simple_plot_2d(ins, res.paths, output_size=4)
```
