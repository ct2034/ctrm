# Planning with Learned CTRMs

```python
import pickle

from ctrm.roadmap_learned import get_timed_roadmaps_multiple_paths_with_learned_indicator
from ctrm.planner import PrioritizedPlanning as PP
from ctrm.utils import set_global_seeds
from ctrm.viz import simple_plot_2d
```

```python
ins = pickle.load(open("/workspace/notebooks/CTRM_demo_ins.pkl", "rb"))
```

```python
set_global_seeds(0)
pred_basename = "/workspace/trained_models/aamas22-main/best"
N_traj = 25
trms = get_timed_roadmaps_multiple_paths_with_learned_indicator(ins, pred_basename, N_traj=N_traj, verbose=1)
planner = PP(ins, trms, time_limit=300, verbose=1)
res = planner.solve()

print(f"elapsed={res.elapsed_planner}")
print(f"makespan={res.maximum_costs}")
print(f"sum-of-costs (normalized)={res.sum_of_costs / ins.num_agents}")
```

```python
simple_plot_2d(ins, res.paths, output_size=6)
```
